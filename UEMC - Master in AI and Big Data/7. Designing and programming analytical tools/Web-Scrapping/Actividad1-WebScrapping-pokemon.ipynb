### Consulta de los datos

```python
r = requests.get('https://pokemondb.net/pokedex/all')
```

```python
page_url = 'https://pokemondb.net/pokedex/all'
html = requests.get(page_url, timeout = 5, verify = True)
print(html)
```

```python
html.text[:1000]
```

```python
dfs = pd.read_html(html.text)
dfs
```

```python
len(dfs)
```

```python
df = dfs[0]
df.head()
```

```python
df.to_csv("pokemon.csv", index = False)
```

### Número de pokemon por tipo

```python
numero_pokemon_tipo = pokemon['Type'].value_counts()
numero_pokemon_tipo
```

```python
numero_pokemon_tipo.to_csv('lista_tipos_pokemon.csv')
```

```python
import pickle

lista_tipos_pokemon = pokemon['Type'].value_counts()
print(type(lista_tipos_pokemon))
print(lista_tipos_pokemon)
with open('lista_tipos_pokemon.txt', 'wb') as file:
    pickle.dump(lista_tipos_pokemon, file)
```

### Mejores pokemon de ataque

```python
pokemon_ordenados_ataque = pokemon.sort_values('Attack', ascending=False)
pokemon_ordenados_ataque
```

```python
pokemon_ordenados_ataque.head(10).to_csv("pokemon_ordenados_ataque.csv")
```

### Mejores pokemon de defensa

```python
pokemon_ordenados_defensa = pokemon.sort_values('Defense', ascending=False)
pokemon_ordenados_defensa
```

```python
pokemon_ordenados_defensa.head(10).to_csv("pokemon_ordenados_defensa.csv")
```
